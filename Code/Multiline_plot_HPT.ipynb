```python
import pandas as pd
path = "/content/drive/Shareddrives/2:1 Caitlin & Kimai/Data/Enzyme Stability Prediction/Experiment Results/HPT - C&K - Sheet1.csv"
data = pd.read_csv(path)
data.head()
```

```python
import matplotlib.pyplot as plt

epochs = list(data["Epochs"])
learning_rate = list(data.columns)
learning_rate = learning_rate[1:]
all_error = []
for index in range(len(epochs)):
  all_error.append(list(data.iloc[index,1:]))

plt.figure(figsize = (15,15))
for index in range(len(epochs)):
  plt.plot(learning_rate, all_error[index], label = "epochs: {}".format(epochs[index]))
plt.xlabel("Learning Rate")
plt.ylabel("Mean Absolute Error")
plt.title("Mean Absolute Error values with respective to Learning rate and Epochs")
plt.grid(True)
plt.legend(loc = "best")
plt.show()
```
